# **Data loading**

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/ITC club/Ai/Task1/Popularity of Programming Languages from 2004 to 2024.csv")
```

# **Data exploration**

```python
print("DataFrame Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nFirst 10 Rows:\n", df.head(10))
print("\nDescriptive Statistics:\n", df.describe())
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
print("\nUnique Values per Column:\n", df.nunique())
print("\nColumn Names:", df.columns.tolist())
```

# **Data cleaning**

```python
# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'], format='%B %Y')

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())

# Check for duplicated rows
print("\nDuplicate Rows:", df.duplicated().sum())

# Verify data types
print("\nData Types:\n", df.dtypes)

# Display descriptive statistics
print("\nDescriptive Statistics:\n", df.describe())
```

# **Data analysis**

```python
# Calculate the average popularity for each language
average_popularity = df.drop('Date', axis=1).mean()

# Identify the top 5 most popular languages
top_5_languages = average_popularity.nlargest(5)

# Calculate the largest popularity increase for each language
largest_increase = {}
for language in df.columns[1:]:
    diff = df[language].diff()
    largest_increase[language] = diff.max()

# Calculate the average popularity growth per year for each language
average_growth_per_year = {}
for language in df.columns[1:]:
    years = len(df) / 12
    total_growth = df[language].iloc[-1] - df[language].iloc[0]
    average_growth_per_year[language] = total_growth / years

# Report the findings
print("Average Popularity:\n", average_popularity)
print("\nTop 5 Most Popular Languages:\n", top_5_languages)
print("\nLargest Popularity Increase:\n", largest_increase)
print("\nAverage Popularity Growth Per Year:\n", average_growth_per_year)
```

# **Data visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a new DataFrame with 'Date' and the top 5 languages
top_5_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C/C++']
df_top_5 = df[['Date'] + top_5_languages].copy()

# Set 'Date' as the index
df_top_5.set_index('Date', inplace=True)

# Create the line plot
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_top_5, palette='tab10', linewidth=2.5)
plt.title('Popularity Trends of Top 5 Programming Languages', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Popularity', fontsize=12)
plt.legend(title='Languages', loc='upper left', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

* Java leads in average popularity with a score of 24.158, followed by Python at 14.726 and PHP at 12.512.
* The top 5 most popular programming languages are Java, Python, PHP, JavaScript, and C/C++.
* Python showed the most significant single-month popularity increase, with a rise of 0.67.
* Python also leads in average annual growth, with a rate of 1.327 per year.
Several languages, including Java, JavaScript, PHP, Perl, Delphi/Pascal, C/C++, Visual Basic, Cobol, Haskell, Matlab, and VBA, experienced negative average annual growth.
